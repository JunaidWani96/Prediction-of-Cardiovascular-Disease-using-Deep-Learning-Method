### Spliting the dataset into taining and testing 

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.01 , random_state=42)
```

```python
def create_model():
    ann = Sequential()
    ann.add(Dense(units=128 , activation='relu',kernel_initializer='random_normal',
    bias_initializer='zeros',input_dim = 11))
    ann.add(Dense(units=128 , activation='relu',kernel_initializer='random_normal',
    bias_initializer='zeros',input_dim = 11))
    ann.add(Dense(units=1 , activation='sigmoid',kernel_initializer='random_normal',
    bias_initializer='zeros'))
    ann.compile(optimizer='adam', loss = 'binary_crossentropy',metrics = ['accuracy'])
    return ann
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
ann=KerasClassifier(model=create_model)
```

```python
parameters = {'batch_size':[100,128] , 'epochs':[100 , 150] , 'optimizer':['adam','rmsprop']}
```

```python
grid_search=GridSearchCV(estimator=ann , param_grid=parameters , scoring='accuracy' ,cv=10)
```

```python
gs = grid_search.fit(X_train , y_train)
best_para = gs.best_params_
best_accu =gs.best_score_
```
